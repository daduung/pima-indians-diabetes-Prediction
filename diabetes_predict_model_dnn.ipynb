```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
```

```python
#csv파일 불러오기
diabetes = pd.read_csv("diabetes.csv")
diabetes
```

```python
#상관관계 분석
diabetes.corr()
```

```python
# 상관관계가 제일 낮은 요소 제거 -> 결과값 확인해보니 예측이 더 떨어짐.. 그래서 주석처리
#del diabetes['BloodPressure']
#diabetes
```

```python
data = X = diabetes.iloc[:,:-1].values
data
```

```python
data.shape
```

```python
label = y = diabetes.iloc[:,-1]
label.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, label,  test_size=0.2, random_state=11)
```

```python
# normalize
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_s = scaler.transform(X_train)
X_train_s
```

```python
scaler = MinMaxScaler()
scaler.fit(X_test)
X_test_s = scaler.transform(X_test)
X_test_s
```

```python
# 결과값이 1개 - sigmoid / 하나 이상은 softmax
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# 4. 모델 컴파일
model.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# X_test, y_test 결과값이 나와야함
history = model.fit(X_train_s, y_train, 
                    validation_data=(X_test_s,y_test), 
                    steps_per_epoch = 30, 
                    validation_steps=50,
                    epochs=30, verbose=1)
```

```python
model.summary()
```

```python
score = model.evaluate(X_train_s, y_train, verbose=1)
print('정답률 = ', score[1],'loss=', score[0])
```
